```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,200)
y1=np.cos(x)
y2=np.sin(x)
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
ax=plt.gca()
ax.set_aspect("equal");
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
ax=plt.gca()
ax.set_aspect("equal")
plt.xlim(0, 2*np.pi);
plt.yticks([-1,0,1], ["-1", "0", "1"]);
plt.xticks([u for u in np.arange(0,2.5*np.pi,np.pi/2)],
[0, r"$\pi$/2", r"$\pi$", r"3$\pi$/2", r"2$\pi$"]);
```

```python
plt.plot(x,y1,ls="-", color="k", label="cosine")
plt.plot(x,y2,ls="--",color="k", label="sine")
ax=plt.gca()
ax.set_aspect("equal")
plt.xlim(0, 2*np.pi);
plt.yticks([-1,0,1], ["-1", "0", "1"]);
plt.xticks([u for u in np.arange(0,2.5*np.pi, np.pi/2)],
[0, r"$\pi$/2", r"$\pi$", r"3$\pi$/2", r"2$\pi$"]);
plt.legend(loc=(.55,.65), fontsize="small")
plt.xlabel("angle in radians")
plt.ylabel("function value")
plt.title("Plots of the Sine and Cosine");
```

```python
# define values on x axis
xvals = np.linspace(0,10,25)
# define values on parabola
y = 10*xvals - xvals**2
# plot the y values vs the x values on a curve
plt.plot(xvals, y, c="black")
# add in some noise
noise = 10*(np.random.rand(25)-.5)
ynoise = y + noise
# plot markers for each noisy data point
# "D" means Diamond shaped
# c means color
plt.scatter(xvals, ynoise, marker="D", c="black")
```

```python
from scipy.special import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
xvals=np.linspace(-15,3.0,200)
airyvals=np.array([airy(x) for x in xvals])
ai = (airyvals.T)[0]
bi = (airyvals.T)[2]
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True)
ax1.plot(xvals, ai, color="black")
ax1.set_ylim(-.75, .75)
ax1.set_title("Airy Ai Function")
ax1.axhline(0.0,color="black")
ax1.axvline(0.0,color="black")
ax2.plot(xvals, bi, color="black")
ax2.set_ylim(-.75, .75)
ax2.set_title("Airy Bi Function")
ax2.axhline(0.0,color="black")
ax2.axvline(0.0,color="black")
```

```python
from scipy.special import *
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# use same x values for all plots
xvals=np.linspace(-1, 1, 200)
# define a 2 by 3 grid of plots that share axes
fig,AX = plt.subplots(nrows=2,ncols=3,sharey=True,sharex=True)
i=1
# Evaluate the Legendre polynomials on x array using scipy
legendre=np.array([list(lpn(6,x)[0]) for x in xvals]).T
# Plot P_i(x) on the ith plot
for row in AX:
    for ax in row:
        yvals=legendre[i]
        ax.plot(xvals,yvals, color="black")
        ax.axhline(0.0, color="black")
        ax.axvline(0.0, color="black")
        ax.set_title("Legendre P"+str(i))
        i+=1
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as colors
#get a list of the colors in alphabetical order
allcolors=list(colors.cnames)
allcolors.sort()
ncolors = len(allcolors) # there are 147
rows = ncolors//3
i=0
fig,ax=plt.subplots(1,1) # 147 is divisible by 3
#
# for each named color, draw a small rectangle
# and put its name in the center
#
for c in allcolors:
    y = rows- (i % rows) - 1
    x = i // rows
    ax.add_patch(patches.Rectangle((x,y), 1, 1, color=c))
    plt.annotate(c, xy=(x+.5,y+.5), ha="center", va="center",
        fontsize=8)
    i+=1
# turn off the ticks
plt.xticks([])
plt.yticks([])
plt.axis([0,3,0,rows])
fig=plt.gcf()
fig.set_size_inches(8,12)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
cmaps = ['Blues', 'BuGn', 'BuPu', 'GnBu', 'Greens', 'Greys',
'Oranges', 'OrRd','PuBu','PuBuGn','PuRd','Purples',
'RdPu','Reds', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd',
'afmhot', 'autumn', 'bone', 'cool', 'copper',
'gist_heat', 'gray', 'hot', 'pink', 'spring',
'summer','winter','BrBG', 'bwr', 'coolwarm','PiYG',
'PRGn', 'PuOr', 'RdBu', 'RdGy', 'RdYlBu', 'RdYlGn',
'Spectral', 'seismic','Accent', 'Dark2', 'Paired',
'Pastel1','Pastel2', 'Set1', 'Set2', 'Set3',
'gist_earth','terrain','ocean','gist_stern','brg',
'CMRmap', 'cubehelix','gnuplot', 'gnuplot2',
'gist_ncar','nipy_spectral', 'jet', 'rainbow',
'gist_rainbow', 'hsv','flag', 'prism']
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))
def plotColorMap():
    fig, axes = plt.subplots(nrows=len(cmaps))
    fig.subplots_adjust(top=0.95, bottom=0.01,
    left=0.2, right=0.99)
    for ax, name in zip(axes, cmaps):
        ax.imshow(gradient, aspect='auto',
            cmap=plt.get_cmap(name))
        pos = list(ax.get_position().bounds)
        x = pos[0] - 0.01
        y = pos[1] + pos[3]/2.
        fig.text(x, y, name, va='center', ha='right',fontsize=10)
    for ax in axes:
        ax.set_axis_off()
    fig.set_size_inches(8,16)
```

```python
plotColorMap()
```
