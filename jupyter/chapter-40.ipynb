```python
import matplotlib.pyplot as plt
%matplotlib inline
def f(n):
    return 0.8*n+1
n=0
y=[n]
x=range(0,15)
for i in x[1:]:
    n = f(n)
    y.append(n)
plt.scatter(x,y)
plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def cobweb(f,x0,N=25):
    x=x0
    for j in range(0,N,2):
        xnew=f(x)
        xold=x
        x=xnew
        plt.plot([xold,x,x],[xnew,xnew,f(xnew)],c="k")
```

```python
cobweb (f,.3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def cobweb_list(f,x0,N=25):
    x=x0; ynext=f(x)
    xv=[]; yv=[]
    for j in range(0,N,2):
        xnew=ynext 
        xold=x
        x=xnew
        ynext=f(x)
        xv=xv+[xold,x,x]
        yv=yv+[xnew,xnew,ynext]
    plt.plot(xv,yv,label="cobweb")
    plt.plot([min(xv),max(xv)],[min(xv),max(xv)],label="y=x")
    xr=np.linspace(min(xv),max(xv),100)
    yr=f(xr)
    plt.plot(xr,yr, label="f(x)")
    plt.legend()
```

```python
cobweb_list(f,.3,25)
```
