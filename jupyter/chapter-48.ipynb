```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
m = Basemap(projection="vandg", lat_0=40,lon_0=-118)

# Longitude, Latitudes
Manilla=(120.9667,14.5833); Johannesburg=(28.0456, -26.2044)
LA=(-118.25, 34.05);        London=(0.08, 51.53)
Sydney=(151.2094, -33.86);  Tokyo=(139.6917, 35.6895)
Santiago=(-70.6667, -33.45)
```

```python
m.drawgreatcircle(LA[0], LA[1], London[0], London[1])
m.drawgreatcircle(LA[0], LA[1],Manilla[0],Manilla[1])
m.drawgreatcircle(Sydney[0], Sydney[1], LA[0], LA[1])
m.drawgreatcircle(LA[0], LA[1], Tokyo[0], Tokyo[1])
m.drawgreatcircle(LA[0], LA[1],Johannesburg[0],
  Johannesburg[1])
m.drawgreatcircle(LA[0], LA[1], Santiago[0], Santiago[1])
m.drawcoastlines()
m.fillcontinents()
m.drawparallels(np.arange(-90,90,10))
m.drawmeridians(np.arange(-180,180,30))
```

```python
m = Basemap(llcrnrlon=-120.,llcrnrlat=20.,\
            urcrnrlon=-45.,urcrnrlat=60.,\
            resolution='l',projection='poly',\
            lat_0=40.,lon_0=-75. )
LA=(-118.25, 34.05);   NY=(-74.0059,40.7129)
m.drawgreatcircle(LA[0],LA[1], NY[0],NY[1],lw=2,color="red")
m.drawcoastlines()
m.drawmapboundary(fill_color="lightblue")
m.fillcontinents(color="white", lake_color="lightblue")
m.drawcountries();  m.drawstates()
m.drawrivers(color="blue")
m.drawparallels(np.arange(20,60,5), linewidth=.25)
m.drawmeridians(np.arange(-120,-45,5), linewidth=.25)
plt.title('Polyconic Projection', fontsize="9")
```
