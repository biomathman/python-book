```python
from math import sqrt
```

```python
def inscribed_polygon(n, echo=False):
    H=sqrt(2)
    p=2
    for i in range(2,n+1):
        H=sqrt(2-2*sqrt(1-(H/2)**2))
        p=2*p
        pii=p*H
        if echo:print(i, pii)
    return pii
```

```python
inscribed_polygon(25,echo=True)
```

```python
def Archimedes(n):
    C=3*sqrt(3)  # circumscribed triangle, perimeter
    I=1.5*sqrt(3) # inscribed triangle, perimeter
    for j in range(n):
        circ=0.5*(C+I)
        print(j,circ)
        Cnext=2*C*I/(C+I)
        Inext=sqrt(Cnext*I)
        
        C=Cnext
        I=Inext
```

```python
Archimedes(15)
```

```python
def taylorpi(n):
    pisum=0
    for j in range(n):
        pisum+=((-1)**j)/(2*j+1)
        
    return(4*pisum)
```

```python
taylorpi(50000)
```

```python
def f(x,n):
    if n>=1:
        mult = -1.0 if n%2==1 else 1.0
        last = (mult)*(x**(2.0*n+1))/(2.0*n+1.0)
        return f(x,n-1)+ last
    else:
        return x

def F(n):
    return 4*(4*f(.2,n)-f(1.0/239.0,n))

for n in range(10):
    print(n, F(n))
```

```python
def spigot():
    q,r,t,k,n,l = 1,0,1,1,3,3
    while True:
        if 4*q+r-t < n*t:
            yield n
            q,r,t,k,n,l = (10*q,10*(r-n*t),t,k, (10*(3*q+r))//t-10*n,l)
        else:
            q,r,t,k,n,l = (q*k,(2*q+r)*l,t*l,k+1,
              (q*(7*k+2)+r*l)//(t*l),l+2)
```

```python
p=spigot()
for i in range(2500):
   print(next(p),end="")
```
