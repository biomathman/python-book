```python
import math
def f(y, t):
    return y*math.sin(t)
```

```python
def euler(f, t0,y0,h,tmax):
    t, y = t0, y0 
    td, yd = [t0], [y0]
    while t<tmax:
        y = y + h*f(y, t)
        yd.append(y)
        t=t+h
        td.append(t)
    return(td,yd)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
(times, values)=euler(f,1,1,0.025,4*math.pi)
plt.plot(times,values)
```

```python
(t1, v1)=euler(f,1,1,0.2,4 *math.pi)
(t2, v2)=euler(f,1,1,0.05,4 *math.pi)
(t3, v3)=euler(f,1,1,0.025,4 *math.pi)
plt.plot(t1,v1, label="$h=0.2$", c="k")
plt.plot(t2,v2, label="$h=0.05$", ls="--",c="k")
plt.plot(t3,v3, label="$h=0.025$",ls="-.",c="k")
```

```python
import numpy as np
def f(v,t):
    x,y=v
    xdot = 10.0*y
    ydot = 10.0*(1-x**2)*y-x
    Yprime = np.array([xdot,ydot])
    return Yprime
```

```python
from scipy.integrate import odeint
times = np.linspace(0,10.0,1001)
v0=np.array([1.0,1.0])
solution=odeint(f,v0,times) # define solution times
```

```python
(xvals,yvals)=solution.T
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(xvals,yvals)
ax2.plot(times,xvals, linestyle="--")
ax2.plot(times,yvals)
```

```python
import matplotlib.pyplot as plt
from scipy.integrate import ode
def f(t, y):
    return .009*y+np.cos(5*t+2*y)/(5-np.cos(t+3*y))
solver=ode(f).set_integrator("vode", method="bdf")
solver.set_initial_value(1,1)
t=[1];
y=[1]
```

```python
while solver.successful() and solver.t<15:
    solver.integrate(solver.t+.01)
    t.append(solver.t)
    y.append(solver.y[0])
plt.plot(t,y,color="k")
plt.xlim(1,15)
plt.title(r"$y'=.009y+\frac{\cos(5t+2y)}{5-\cos(t+3y)}$")
```
