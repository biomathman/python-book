```python
import numpy as np
from scipy.interpolate import interp1d
years=range(1900,2000,10)
population=[76.094,92.407,106.461,123.077,132.122, 152.271,
  180.671,205.052,227.224,249.464]
f=interp1d(years, population)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(1900,1990,100)
y=f(x)
plt.plot(x,y,color="Red")
plt.scatter(years, population, color="Black")
plt.axis([1895,1995,50,275])
plt.xticks(range(1900,1995,10), rotation=75)
plt.ylabel("US Population, Millions")
plt.grid()
```

```python
import numpy as np
from scipy.interpolate import splrep,splev
x=np.linspace(1900,1990,100)
f=splrep(years,population)
y=splev(x,f)
plt.plot(x,y,color="Red")
plt.scatter(years,population,color="Black")
plt.xlim(1895,1995)
plt.ylim(50,275)
plt.xticks(range(1900,1995,10), rotation=75)
plt.ylabel("US Population, Millions")
```

```python
import numpy as np
from scipy.interpolate import griddata

# original function
def f(x,y):
    return np.cos(x) + np.sin(y)
    
# define a full grid
xg,yg=np.mgrid[np.pi:3*np.pi:100j, np.pi:3*np.pi:100j]

# get points on the grid
xpts = np.linspace(np.pi, 3*np.pi,10)
ypts = np.linspace(np.pi, 3*np.pi,10)
pts = np.array([[x,y] for y in ypts for x in xpts])       
fpts = f(pts[:,0], pts[:,1])
```

```python
# do a cubic interpolation
fit = griddata(pts, fpts, (xg,yg), method="cubic")

# do a nearest interpolation
fitn = griddata(pts, fpts, (xg,yg), method="nearest")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
# plot original function
p1=ax1.imshow(f(xg,yg).T, 
   extent=[np.pi, 3*np.pi, np.pi, 3*np.pi], origin="lower")
ax1.set_title("Original Function")
# plot cubic interpolation
c2=ax2.imshow(fit.T, extent=[np.pi, 3*np.pi, np.pi, 3*np.pi],
              origin="lower")
ax2.set_title("Cubic Interpolation")
# plot nearest interpolation
c3=ax3.imshow(fitn.T, extent=[np.pi, 3*np.pi, np.pi, 3*np.pi],
              origin="lower")
ax3.set_title("Nearest Interpolation")
```
