```python
import numpy as np
from matplotlib.pyplot import *
%matplotlib inline
```

<h4>Generate random exponential data set

```python
rvals=[0.569550796190182, 0.9345827459164009, 0.1264559247263517, -0.013179954991248577, 0.21953037257965535, -0.763568636073978, -0.08402571582340923, 0.3156930853557032, -0.060835957880793545, -0.32840369668038116, -0.17273595570063116, 0.8903575537124886, 0.7745395166885052, 0.4936708925938311, 0.25601526125678764, 0.7804558310595537, 0.3155481551979582, 0.3548967374447234, -0.9868432657639221, -0.6314573808899537, -0.08480763466013164, -0.1898164121361623, -0.98312244406877, 0.15552595072775133, -0.8424079778476778, 0.6211793987799112]
```

```python
xvals=np.linspace(-4,2,400)
#rvals=[.5*(np.random.rand()-.5) for x in xvals]

yvals=np.exp(.075*xvals+.001)+rvals
exactvals=np.exp(0.075*xvals+0.001)
scatter(xvals,yvals, s=.1)
plot(xvals,exactvals)

xlabel("x")
ylabel("y")
fig=gcf()
fig.set_size_inches(6,4)
fig.savefig("Randomized-Data-About-Exponential.pdf",dpi=300)
```

<h4>Least Squares fit of Log y vs. x

```python
logy=np.log(yvals)
linearfit=np.polyfit(xvals,logy,1)
scatter(xvals,logy, s=.1)
thefit=[np.polyval(linearfit, x) for x in xvals]
plot(xvals, thefit)
xlabel("Original x value")
ylabel("ln(y)")
fig=gcf()
fig.set_size_inches(6,4)
fig.savefig("linear-regression-in-log-space.pdf",dpi=300)

print linearfit
```

```python
m,b=linearfit;
print m, b
def predicted(x):
    return np.exp(m*x+b)
```

```python
pdata=[predicted(x) for x in xvals]
plot(xvals,pdata, c="red", label="least squares log")
scatter(xvals,yvals, s=.1, c="green")
plot(xvals,exactvals, c="blue", label="exact data")
legend(loc="best")
xlabel("x value")
ylabel("y value")
fig=gcf()
fig.set_size_inches(6,4)
fig.savefig("exponential-log-regression.pdf",dpi=300)
```

<h5>Figure out least squares landscape

```python
def rss(m, b):
    def pred(x,m,b):
        return np.exp(m*x+b)
    preds=[pred(x,m,b) for x in xvals]
    rss = sum([(y1-y2)**2 for y1,y2 in zip(preds, yvals)])
    return np.sqrt(rss)
```

```python
import numpy as np
from matplotlib.pyplot import *
%matplotlib inline

USPOP=[3929, 5308,7239,9638,12866,
       17069,23191,31433,38558,50189,
       62979,76212,92228,106021,123202,
       132164, 151325,179323,203302,226542,248709]
USYR =[1790,1800,1810,1820,1830,
       1840,1850, 1860,1870,1880,
       1890,1900,1910,1920,1930,
       1940,1950,1960,1970,1980,1990]
xvals=[X-1790 for X in USYR]
scatter(xvals, USPOP,label="Actual Population")
logpop=np.log(USPOP)
linearfit=np.polyfit(xvals,logpop,1)

m,b=linearfit;
print m, b
def predicted(x):
    return np.exp(m*x+b)
pdata=[predicted(x) for x in xvals]
plot(xvals,pdata, c="red", label="Least Squares Log")
xlabel("Year")
ylabel("US Population (thousands)")
xticks(range(10,201,50), range(1800,1991,50))
xlim(-5,205)
ylim(-20000,400000)
legend(loc="best")
fig=gcf()
fig.set_size_inches(6,4)
fig.savefig("USPOP-LSEXP-FIRSTFIT.pdf",dpi=300)
```

```python
def rmspop(m, b):
    def pred(x,m,b):
        return np.exp(m*x+b)
    preds=[pred(x,m,b) for x in xvals]
    rss = sum([(y1-y2)**2 for y1,y2 in zip(preds, USPOP)])
    return np.sqrt(rss)
```

```python
xpoint=25*(.0207-.0)/(.04-0); ypoint=200*(8.617-1)/(15-1)
(xpoint, ypoint)

PLUSXpoint=25*(.015-.0)/(.04-0); PLUSYpoint=200*(9.5-1)/(15-1)
```

```python
A=[[rmspop(M,B) for M in np.linspace(0.0,.04,25)] 
   for B in np.linspace(1,15,200)]
cp=contourf(A, levels=np.linspace(0,700000,15))

colorbar(cp)
scatter([xpoint],[ypoint], label="Logarithmic Least Squares", 
       s=75, facecolor="pink")
scatter([PLUSXpoint],[PLUSYpoint],marker="D", s=75,
        facecolor="white", label="Landscape Minimum")
xticks(range(0, 25, 6), np.linspace(0.0, 0.04,5))
#yticks(range(0, 200,49 ), np.linspace(1, 15,5))
yticks([(199./14.)*(x-1) for x in range(6,13)], range(6,13))
ylim((199./14.)*5., (199./14.)*12)
legend(loc="upper right")
xlabel("m");ylabel("b")
fig=gcf()
fig.set_size_inches(6,6)
fig.savefig("Population-Landscape-Minimum.pdf",dpi=300)
```
