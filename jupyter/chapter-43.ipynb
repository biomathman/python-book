```python
from PIL import Image
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
g=Image.open("flower.jpg") # read the image
g=g.convert("L") # convert to gray scale
A=np.asarray(g) # convert to an array
U, Sigma, V = np.linalg.svd(A)
UT=U.T  # transpose 
M = np.zeros_like(A)
for j in range(120):  # Compute matrix product
    B=np.outer(UT[j], V[j]) * Sigma[j]
    M=M+B
i=Image.fromarray(M) # convert to image
#i.show() # to show the picture in a pop-up
plt.imshow(i)
```
